## Análisis de los null

### Columna CONSTRUCTIONYEAR

```python
df[df["CONSTRUCTIONYEAR"].isnull()][:10]
```

Hay una columna que tiene los mismos datos que CONSTRUCTIONYEAR. Esta columna se llama CADCONSTRUCTIONYEAR y podemos usarla para rellenar los nulos de CONSTRUCTIONYEAR.

```python
df[["CONSTRUCTIONYEAR", "CADCONSTRUCTIONYEAR"]][:10]
```

```python
df.loc[df["CONSTRUCTIONYEAR"].isnull(), "CONSTRUCTIONYEAR"] = df[df["CONSTRUCTIONYEAR"].isnull()]["CADCONSTRUCTIONYEAR"]
```

```python
df["CONSTRUCTIONYEAR"].isnull().sum()
```

La columna CONSTRUCTIONYEAR es el año de construcción dada por el anunciante de la propiedad. La columna CADCONSTRUCTIONYEAR es la fecha de construcción catastral. Aunque puedan no ser la misma fecha, no podemos inferir las fechas de construcción de otras columnas. Es por eso que una buena primera aproximación sería asignar a los null en CONSTRUCTIONYEAR su correspondiente CADCONSTRUCTIONYEAR.

### Columna FLOORCLEAN

FLOORCLEAN indica el número de la planta en la que se encuentra la propiedad. En este caso no podemos inferir todos los null, pero podemos inferir los que se encuentren en la planta más alta asignándoles la planta más alta de su edificio.

```python
df["ISINTOPFLOOR"].value_counts()
```

```python
df.loc[(df["FLOORCLEAN"].isnull()) & (df["ISINTOPFLOOR"]==1), "FLOORCLEAN"] = df.loc[(df["FLOORCLEAN"].isnull()) & (df["ISINTOPFLOOR"]==1), "CADMAXBUILDINGFLOOR"]
```

```python
df["FLOORCLEAN"].isnull().sum()
```

Solo hemos podido reducir en 13 los nulls.

También podemos rellenar las propiedades en las que el máximo de plantas sea 1 y la propiedad no esté en la planta más alta. En este caso la planta es la 0.

```python
df.loc[(df["FLOORCLEAN"].isnull()) & (df["ISINTOPFLOOR"]==0) & (df["CADMAXBUILDINGFLOOR"]==1.0), "FLOORCLEAN"] = 0
```

```python
df["FLOORCLEAN"].isnull().sum()
```

# Estudio de precios

Para saber qué columnas influyen más en los precios generamos una matriz de correlación.

```python
df.corr().loc["PRICE"]
```

El coeficiente de correlación de Pearson se interpreta de la siguiente manera:

|valores|fuerza de la correlación|dirección de la correlación|
|---|---|---|
|entre 0.5 y 1|fuerte|positiva|
|entre 0 y 0.5|débil|positiva|
|0|neutra|neutra|
|entre 0 y -0.5|débil|negativa|
|entre -0.5 y -1|fuerte|negativa|

Teniendo esta tabla en cuenta, podemos decir que las columnas que más correlación tienen con la columna PRICE son BATHNUMBER, CONSTRUCTEDAREA, CADASTRALQUALITYID, DISTANCE_TO_CITY_CENTER

## Precio VS Área construída

```python
for i in np.arange(0, 2772000 , 250000, ):
  df_price_area = df.loc[(df["PRICE"] > i) & (df["PRICE"] < i+250000)]
  x = df_price_area["CONSTRUCTEDAREA"]
  y = df_price_area["PRICE"]

  plt.figure(figsize=(15,10))
  plt.xlabel("Constructed Area")
  plt.ylabel("Price")
  # plt.yticks(np.arange(0, 2772000 , 250000, ))
  plt.scatter(x, y) #s=area, c=colors, alpha=0.5)
  plt.show()
```

A precio más alto, apreciamos que los valores de área construida son mayores.

## Precio vs número de baños

```python
for i in np.arange(0, 2772000 , 250000, ):
  df_price_area = df.loc[(df["PRICE"] > i) & (df["PRICE"] < i+250000)]
  x = df_price_area["BATHNUMBER"]
  y = df_price_area["PRICE"]

  plt.figure(figsize=(15,10))
  plt.xlabel("Number of bathrooms")
  plt.ylabel("Price")
  plt.scatter(x, y)
  plt.show()
```

A precios más altos, podemos apreciar que es más frecuente un número mayor de cuartos de baño.

## Precio vs CADASTRALQUALITYID

La calidad catastral es la calidad de construcción del edificio o vivienda:

https://www.juntadeandalucia.es/institutodeestadisticaycartografia/badea/metainformacion?CodOper=b3_151&idMetainf=Bienes_inmuebles_seg%C3%BAn_calidad_constructiva

La máxima calidad es el valor 1.

```python
for i in np.arange(0, 2772000 , 250000, ):
  df_price_area = df.loc[(df["PRICE"] > i) & (df["PRICE"] < i+250000)]
  x = df_price_area["CADASTRALQUALITYID"]
  y = df_price_area["PRICE"]

  plt.figure(figsize=(15,10))
  plt.xlabel("Cadastral quality")
  plt.ylabel("Price")
  plt.scatter(x, y)
  plt.show()
```

Evidentemente, a mayor calidad catastral, mayor es el precio.

## Precio vs DISTANCE_TO_CITY_CENTER

```python
for i in np.arange(0, 2772000 , 250000, ):
  df_price_area = df.loc[(df["PRICE"] > i) & (df["PRICE"] < i+250000)]
  x = df_price_area["DISTANCE_TO_CITY_CENTER"]
  y = df_price_area["PRICE"]

  plt.figure(figsize=(15,10))
  plt.xlabel("Distance to city center")
  plt.ylabel("Price")
  plt.scatter(x, y)
  plt.show()
```

Por regla general, los inmuebles más cercanos al centro son más caros. Hay excepciones. Lo más probable es que estas excepciones sean debidas a propiedades en las afueras con gran extensión de terreno.
